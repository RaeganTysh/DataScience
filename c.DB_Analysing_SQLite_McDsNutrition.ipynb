## Analysing McDonalds Nutritional information

```python
import pandas as pd
import sqlite3
data=pd.read_csv('./menu.csv')
conn= sqlite3.connect('Mcdonalds.db')
data.to_sql('MCDONALDS_NUTRITION', conn)
```

```python
df= pd.read_sql("SELECT * FROM MCDONALDS_NUTRITION", conn)
print(df)
```

```python
df.head()
```

```python
#common mathematical and statistical methods - statistical method
df.describe(include ='all')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#categorical scatterplots
plt.figure(figsize=(10,5)) #size x,y
plot =sns.swarmplot (x="Category", y="Sodium", data=df, size=2) #anything greater than 3 cannot account for all data - try it out
plt.setp(plot.get_xticklabels(), rotation=70)
plt.title('Sodium Content')
plt.show()
```

```python
#stripplot allows for overlapping dat point(can obscure data interp)- best ofr smaller datasets
plot =sns.stripplot (x="Category", y="Sodium", data=df)
plt.setp(plot.get_xticklabels(), rotation=70)
plt.title('Sodium Content')
plt.show()
```

```python
import pandas as pd        #data manipulation
import matplotlib.pyplot as plt    #additional plotting capabilities
%matplotlib inline
import seaborn as sns      #plotting
from scipy.stats import pearsonr  #statistical calculations

# Calculate Pearson correlation coefficient and p-value
corr_coef, p_value = pearsonr(df['Protein'], df['Total Fat'])

plot = sns.jointplot(x="Protein", y="Total Fat", data=df, size=3, kind='scatter', color='purple')

# Annotate the plot with the Pearson correlation coefficient and p-value
plt.annotate(f'Pearson r: {corr_coef:.2f}\np-value: {p_value:.2e}', 
             xy=(0.5, 0.9), xycoords='axes fraction', 
             fontsize=12, ha='center', 
             bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))
plt.show()
```

## Box: The box represents the middle 50% of the data, with the bottom of the box at Q1 and the top at Q3. The horizontal line inside the box represents the median (Q2).
## Whiskers: The whiskers extend from the ends of the box to the minimum and maximum values that are within 1.5 times the interquartile range (IQR = Q3 - Q1) from the ends of the box. Values beyond the whiskers are considered potential outliers.
## Outliers: Points that lie beyond the whiskers are plotted individually and are considered potential outliers.

```python
#ex box plot
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create the boxplot
sns.boxplot(x="species", y="sepal_length", data=iris)

plt.title("Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

## This code will generate a boxplot that shows the distribution of sepal length for each species of iris flower. The x-axis represents the different species, while the y-axis shows the sepal length values.
## Boxplots are useful for comparing the distributions of multiple datasets or groups, as they provide a quick visual summary of the data. They are particularly helpful in identifying potential outliers and comparing the medians and quartiles of different groups.
