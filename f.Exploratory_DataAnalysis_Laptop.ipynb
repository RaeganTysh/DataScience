# Task 3 - GroupBy and Pivot Tables

Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.


````python
# Write your code below and press Shift+Enter to execute
# Create the group in new df
df_gptest = df[['GPU','CPU_core','Price']]
#create new df as group by
grouped_test1 = df_gptest.groupby(['GPU','CPU_core'],as_index=False).mean()
print(grouped_test1)
````

````python
# Write your code below and press Shift+Enter to execute
# Create the Plot
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#title and axis
plt.title("Pivot Plot CPU-Core vs GPU", fontsize=16)

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

# Add X and Y axis labels
ax.set_xlabel('CPU-Core', fontsize=12)
ax.set_ylabel('GPU', fontsize=12)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

fig.colorbar(im)

# Adjust layout to prevent clipping of tick-labels
plt.tight_layout()

plt.show()
````

# Task 4 - Pearson Correlation and p-values

Use the `scipy.stats.pearsonr()` function to evaluate the Pearson Coefficient and the p-values for each parameter tested above. This will help you determine the parameters most likely to have a strong effect on the price of the laptops.


````python
# Write your code below and press Shift+Enter to execute
for param in ['RAM_GB','CPU_frequency','Storage_GB_SSD','Screen_Size_inch','Weight_pounds','CPU_core','OS','GPU','Category']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['Price'])
    print(param)
    print("The Pearson Correlation Coefficient for ",param," is", pearson_coef, " with a P-value of P =", p_value)
````

<details>
<summary>Click here for Solution</summary>


</details>


````python
#Alternatively store the above as a new df
# Create an empty list to store the results
results = []

# Iterate through the parameters and calculate correlations
for param in ['RAM_GB', 'CPU_frequency', 'Storage_GB_SSD', 'Screen_Size_inch', 'Weight_pounds', 'CPU_core', 'OS', 'GPU', 'Category']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['Price'])
    
    # Append the results to the list
    results.append({
        'Parameter': param,
        'Pearson_Correlation': pearson_coef,
        'P_value': p_value
    })

# Create a DataFrame from the results list
correlation_df = pd.DataFrame(results)

# Sort the DataFrame by absolute correlation value (descending order)
correlation_df['Abs_Correlation'] = correlation_df['Pearson_Correlation'].abs()
correlation_df = correlation_df.sort_values('Abs_Correlation', ascending=False).drop('Abs_Correlation', axis=1)

# Display the DataFrame
print(correlation_df)

# Optionally, you can save this to a CSV file
# correlation_df.to_csv('correlation_results.csv', index=False)
````

# Now we can use the following visualizations to easily interpret the relationships between different parameters and the price

````python
#make a bar plot - with matplotlib
plt.figure(figsize=(12, 6))
plt.bar(correlation_df['Parameter'], correlation_df['Pearson_Correlation'])
plt.title('Pearson Correlation Coefficients with Price')
plt.xlabel('Parameters')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````
